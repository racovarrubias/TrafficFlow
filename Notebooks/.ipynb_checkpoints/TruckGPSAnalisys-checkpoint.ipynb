```python
import pandas as pd
import json
import csv, sys
import os
import os.path
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#read data filename
root = 'C:/Users/rc12345/Traffic Champaign/Projects/GPSData/'
filename = 'CHAMP_APR2017.csv'
full_path = os.path.join(root, filename)

flowApril = pd.read_csv(full_path, names=['Date', 'Latitude', 'Longitude', 'Speed','TruckID'], low_memory=False, skiprows=1)

flowApril.head(10)
```

```python
#Drop  all rows that do not contain latitude or longitude

cols_drop = ['Latitude', 'Longitude','TruckID']
clean_flowApril = flowApril.dropna(subset = cols_drop)
print(clean_flowApril.shape, flowApril.shape)

clean_flowApril.head(5)
```

```python
#transform Date data to datetime
clean_flowApril['Date'] = pd.to_datetime(clean_flowApril['Date'])
#add a colukn with weekday name DayWeek
clean_flowApril['Day of Week'] = clean_flowApril['Date'].dt.weekday_name
#clean_flowApril = clean_flowApril.groupby(['DayWeek'])
clean_flowApril.head(5)
```

Determine the distribution of observations per day of the week. We would like to know if there are some non-uniformities in the number of datapoints taken.

To do this, group by day of the week and count the number of observations per day.


```python
#total number of observations
total_observations = clean_flowApril.shape[0]

cats = [ 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

#group by day of the week and count all the observations for each day
observations_per_day = clean_flowApril.groupby(['Day of Week']).Speed.count().reindex(cats)

#group by day of the week and unique truck ID and counts number of observations per day per truck
observations_per_truck_day = clean_flowApril.groupby(['Day of Week', 'TruckID']).count()

observations_per_truck_day.Speed.plot.hist(bins=40)
```

```python
#Plot each day on a bar plot
#Let's transform the values to percentage 
observations_per_day = 100 * observations_per_day / total_observations

```

```python
#plot the data in a bar plot. 
ax = observations_per_day.plot(kind='bar', colormap='Paired', title='Observations per day of the week')

#set the y axis label
ax.set_ylabel("Percent")
```

#### How many observations each truck has per day

We would like to know if a truck has too little number of observations in a given day of the week. 

Let's divide the dataset in bins on number of observations a truck has in a day.


```python
bins = [0, 40, 80, 120, 160, 200, 240, 280, 320, 360, 400, 440, 480, 1000]

temp = observations_per_truck_day

temp['binned'] = temp.groupby(pd.cut(temp['Date'], bins)).count()

temp.plot(kind='bar')
#plot the number of observations per truck per day
#ax = temp.plot.bar(rot=0, color='b', figsize=(6,4))
#ax.set_xticklabels([c[1:-1].replace(',', 'to') for c in temp.cat.categories])
#plt.show()
```
